```python
# Importing dependencies for comunicating with sqlite db
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy import create_engine,inspect, func, desc
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import Session

# Importing dependencies for data analysis
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib
from matplotlib import style
style.use('seaborn')
import matplotlib.pyplot as plt
```

```python
# Create engine to connect to hawaii.sqlite database
engine = create_engine('sqlite:///hawaii.sqlite', echo=False)
Base = automap_base()
```

```python
# Using Base.prepare and Base.classes we will reflect DB into ORM classes
Base.prepare(engine, reflect=True)
Base.classes.keys()
```

```python
# Explore database with inspector
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# Save the refference from table mesurements ana station
Measurs = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create db sesion object
session = Session(engine)
```

# Precipitation Analysis
+ Retrieve the last 12 months of precipitation data.
+ Select only the `date` and `prcp` values.
+ Load the query results into a Pandas DataFrame and set the index to the date column.
+ Sort the DataFrame values by `date`.
+ Plot the results using the DataFrame `plot` method.

```python

# Creating the query to identify last date to identify end ponint of the year
ld = session.query(Measurs.date).order_by(Measurs.date.desc()).first()

# print(ld) returns '2017-08-23' which means we will be working with peroiod between 2016-8-23 and 2017-8-23
precip_year = session.query(Measurs.date, Measurs.prcp).filter(Measurs.date >= '2016-8-23').order_by(Measurs.date).all()
```

```python
# Creating percipitation df, reseting the date to datetime format and setting indext to date
precip_df = pd.DataFrame(precip_year, columns = ['Date' , 'Precip'])
precip_df['Date'] = pd.to_datetime(precip_df['Date'], format='%Y/%m/%d')
precip_df.set_index(precip_df['Date'])
precip_df.head()
```

```python
#Plot the results using the DataFrame plot method.
precip_df.plot('Date', 'Precip', figsize=(15, 8))
plt.xlabel("Date")
plt.ylabel("Inches Precipitation")
plt.title("Hawaii Precipitation Analysis 8/23/2016 to 8/23/2017")
plt.legend(["Precipitation"])
plt.savefig("precipitation.png")
#plt.tight_layout()
plt.show()
```

```python
# Printing the summary of stats
precip_df.describe()
```

# Station Analysis
+  Calculate the total number of stations.
+  Find the most active stations.
+  List the stations and observation counts in descending order.
+  Which station has the highest number of observations?
+  Retrieve the last 12 months of temperature observation data (tobs).
+  Filter by the station with the highest number of observations.
+  Plot the results as a histogram with `bins=12`.

```python
# Runing the count of station in Mesrusments table
stations = session.query(Measurs).group_by(Measurs.station).count()
print("There is total of {} stations.".format(stations))
```

```python
# Looking for the most active stations
ma_station = session.query(Measurs.station, func.count(Measurs.tobs)).group_by(Measurs.station).\
               order_by(func.count(Measurs.tobs).desc()).all()

most_active = ma_station[0][0]    
print("The most active station is",most_active,"and has",ma_station[0][1],"weather observations.")
print()

print("List of other stations:")

for station, count in ma_station:
    print("Station",station,"had",count, "weather observations.")
```

```python
# Retriving last 12 months of temp. observ.
temp = session.query(Measurs.station, Measurs.date, Measurs.tobs).\
    filter(Measurs.station == most_active).\
    filter(Measurs.date > '2016-8-23').\
    order_by(Measurs.date).all()
```

```python
# Creating temp data frame and ploting the data
temp_df=pd.DataFrame(temp)
plt.hist(temp_df['tobs'],12)
plt.xlabel("Recorded Temperature")
plt.ylabel("Number of Recorded Observations")
plt.title("Station Analysis (8/23/16 to 8/23/17) for Station " + most_active)
plt.show()
```
